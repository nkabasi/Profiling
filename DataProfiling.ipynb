```
import pandas as pd
import pandas_profiling as pp
import pandas.util.testing as tm
```

```
from pandas_profiling import ProfileReport
```

```
features_data = pd.read_csv("datasets_2296_3883_Features_data_set2.csv",sep=',')
```

```
len(features_data)
```

```
features_data.head()
```

```
features_data.describe()
```

```
features_profile=pp.ProfileReport(features_data,title='Features Profiling Report')
features_profile.to_file("features_profile.html")
```

```
stores_data.head()
```

```
stores_data.describe()
```

```
stores_profile=pp.ProfileReport(stores_data,title='Stores Profiling Report')
stores_profile.to_file("stores_profile.json")
```
